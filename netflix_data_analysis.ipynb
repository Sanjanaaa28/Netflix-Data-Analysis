```python
df = pd.read_csv("netflix_titles.csv")
```

```python

plt.figure(figsize=(8,6))
sns.countplot(data=df, x='type', palette='pastel')
plt.title('Movies vs TV Shows on Netflix')
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar', color='lightgreen')
plt.title("Top 10 Countries with Most Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.xticks(rotation=60)
plt.show()
```

```python
sns.countplot(x='rating', hue='type',data=df, order=df['rating'].value_counts().index)
plt.title("Rating Distribution of Movies and TV Shows on Netflix")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="Content Type")
plt.show()
```

```python
# Content released over the years 
df['release_year'].value_counts().sort_index().plot(kind='line', marker='d', color='skyblue')
plt.title("Content Released Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
# Top 10 genres 
df['listed_in'].str.split(',').explode().value_counts().head(10).plot(kind='bar', color='lightpink')
plt.title("Top 10 Netflix Genres")
plt.xlabel("Genre")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```

```python
# Rating distribution 
df['rating'].value_counts().plot(kind='bar', color='lightgreen')
plt.title("Netflix Content Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```

```python
# Movies vs TV Shows 
type_counts = df['type'].value_counts()

# Pie chart plot
plt.figure(figsize=(8,6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set3"), shadow=True)
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
# Pie chart Rating 
netflix['rating'].value_counts().plot.pie(autopct='%1.1f%%' , shadow=True, figsize=(12,8),colors=sns.color_palette("Set2"))
plt.title("Rating")
plt.show()
```

```python
# Top 5 countries
top_countries = df['country'].value_counts().head(5)

# Pie chart plot
plt.figure(figsize=(7,7))
plt.pie(top_countries, labels=top_countries.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set3")[0:5])
plt.title("Top 5 Countries with Most Netflix Content")
plt.show()
```

```python

genre_counts = df['listed_in'].str.split(',').explode().value_counts().head(5)

# Pie chart plot
plt.figure(figsize=(6,6))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel")[0:5])
plt.title("Top 5 Netflix Genres")
plt.show()
```

```python
df['rating'].value_counts().plot(kind='bar', color='lightgreen')
plt.title("Netflix Content Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```

```python
# Pivot table: type vs rating
pivot_table = df.pivot_table(index='type', columns='rating', aggfunc='size', fill_value=0)

# Heatmap
plt.figure(figsize=(12,4))
sns.heatmap(pivot_table, annot=True)
plt.title("Netflix Content: Type vs Rating")
plt.show()
```
